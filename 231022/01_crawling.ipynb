### 통신 방식
- get
    - 가장 일반적으로 사용하는 통신 방식
    - 웹 브라우저에서 주소창에 주소를 입력하고 요청을 보내는 방식
    - 유저가 서버에게 데이터를 보내는 경우 주소 뒤에 ?로 데이터의 시작을 알리고 key=value의 형태로 데이터를 보낸다.
    - 유저가 입력한 데이터가 주소창에 노출이 됨으로 보안상의 문제가 발생할 수 있다.
    - 속도는 가장 빠르다.
    - 주소창에 데이터를 보내기 때문에 데이터의 길이가 제한(max 255자)
- post
    - get 방식에서 보안적인 측면에서 강화한 데이터 통신 방식
    - 유저가 보내는 데이터를 body라는 부분에 숨겨서 데이터를 보내는 방식
    - 데이터의 길이 제한이 없다.
    - 데이터가 노출이 되지 않기 때문에 보안적으로 뛰어나다.
    - 웹 브라우저에서 일반적인 주소창에 주소를 입력하는 방식으로 접속이 불가능
    - get 방식에 비해 속도가 느리다.

# beautifulsoup 라이브러리
1. html Tag로 이루어진 문자열 데이터를 파싱하여 데이터를 쉽게 추출하기 위한 라이브러리
2. 웹의 구조를 어느 정도 파악하고 사용하면 쉽게 데이터를 추출

```python
# 라이브러리 로드
from bs4 import BeautifulSoup as bs
```

```python
# 파싱 작업 (데이터의 형태를 변환 : beautifulsoup라이브러리에서 사용이 가능한 형태로)
soup = bs(html_data, 'html.parser')
soup
```

```python
print(type(html_data))
print(type(soup))
```

- 태그의 이름을 사용하여 검색
    - soup.태그명 : 해당 태그의 첫번째 정보를 출력
    - soup.태그명.string : 첫번째 태그의 문자열을 출력
    - soup.태그명['속성명'] : 첫번째 태그의 속성의 값을 출력


```python
# title이라는 태그의 정보를 출력
print(soup.title)
print(soup.title.string)
```

```python
print(soup.p)
print(soup.p['class'])
```

- find()
    - html 문자열 안에서 해당 태그의 첫번째 정보를 출력
    - find(속성 = 값) : 태그들 중 해당 속성의 값과 같은 태그의 첫번째 정보를 출력
- find_all()
    - html 문자열 안에서 해당 태그의 모든 정보를 리스트 형태로 출력
    - limit 매개변수 : 리스트의 길이를 지정

```python
soup.find('a')
```

```python
soup.find('a').get_text()
```

```python
# a태그글의 모든 정보를 출력
a_list = soup.find_all('a')
```

```python
soup.find_all('a')[0].get_text()
```

```python
# a태그들의 문자로만 이루어진 배열 데이터를 생성

# find_all()에서는 get_text() 사용 불가능
# 리스트의 각 항목들에서는 get_text() 사용 가능
# 반복문을 이용하여 각 항목들을 대입하여 get_text()를 사용하여 문자열을 새로운 리스트에 추가

# 새로운 리스트 변수를 생성
a_result = []

for i in a_list :
    # i : a_list 각 항목들이 대입
    a_result.append(i.get_text())

a_result
```

```python
# map( 함수(행동) , 자료형데이터 )
list(map(
    lambda x : x.get_text(),
    a_list
))
```

```python
def change(x):
    return x.get_text()

list(map(
    change,
    a_list
))
```

# 네이버쇼핑에서 데이터 크롤링 하기

```python
url = "https://search.shopping.naver.com/search/all?query=아이패드&bt=-1&frm=NVSCPRO"

response = requests.get(url)

# response의 html data만 변수에 대입
html_data = response.text

print(html_data)
```

```python
# beautifulsoup을 이용하여 데이터 파싱
soup = bs(html_data, 'html.parser')
```

```python
div_data = soup.find("div", attrs={
    'class' : 'basicList_list_basis__uNBZx'
})
```

```python
div_data.find('div', attrs={
    'class' : 'adProduct_title__amInq'
    }).get_text()
```

```python
ad_title_list = div_data.find_all('div', attrs={
    'class' : 'adProduct_title__amInq'
    })
```

```python
title = []

for t in ad_title_list :
    title.append(t.get_text())

title
```

```python
ad_price_list = div_data.find_all('span', attrs={
    'class' : 'price_num__S2p_v'
})
ad_price_list
```

```python
price_list = list(map(
    lambda x : x.get_text(),
    ad_price_list
))

price_list
```

```python
# 리스트의 마지막 항목을 삭제
price_list.pop()
price_list
```

```python
ad_img_list = div_data.find_all('img')
```

```python
# img 태그 안에 src 속성 값을 출력

ad_img_list[0]['src']
```

```python
img_list = list(map(
    lambda x : x["src"],
    ad_img_list
))
img_list
```

```python
div_data.find('a', attrs = {
    'class' : 'thumbnail_thumb__Bxb6Z'
})
```
