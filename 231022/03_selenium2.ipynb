```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 웹드라이버를 오픈
driver = webdriver.Chrome()
```

```python
# 웹드라이버에 네이버로 접속
driver.get('http://www.naver.com')
```

```python
# 검색어창 태그를 선택
element = driver.find_element(By.ID, 'query')
```

```python
# 검색어창에 특정 문자열 입력
element.send_keys('트립어드바이저')
```

```python
# 검색어창에서 ENTER키 이벤트 발생
element.send_keys(Keys.ENTER)
```

```python
# class가 link_name인 태그를 선택
element2 = driver.find_element(By.CLASS_NAME, 'link_name')
```

```python
# element2를 클릭 이벤트 발생
element2.click()
```

```python
# 자식창으로 이동
all_windows = driver.window_handles
driver.switch_to.window(all_windows[1])
```

```python
# name이 q인 태그를 선택
element3 = driver.find_element(By.NAME, 'q')
```

```python
# element3에 특정 문자열을 입력
element3.send_keys('치앙마이')
```

```python
element3.send_keys(Keys.ENTER)
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
# div 태그 중 class의 값이 prw_search_search_results인 태그의 정보를 추출
div_data = soup.find('div', attrs={
    'class' : 'prw_search_search_results'
})
div_data
```

```python
div_data2 = div_data.find('div', attrs={
    'class' : 'is-multiline'
})
div_data2
```

```python
# div_data2dㅔ서 class가 result-title인 태그를 모두 찾는다.
title_list = div_data2.find_all('div', attrs={
    'class' : 'result-title'
})

title_list
```

```python
title = []

for i in title_list:
    title.append(i.get_text())
title
```

```python
# 맵 함수 사용
#list(map(
#    lambda x : x.get_text(),
#    title_list
#))
```

```python
# img 주소를 추출하기 위해서 div 태그 중 class가 is-shown-at-desktop인 데이터를 모두 추출
img_list = div_data2.find_all('div', attrs={
    'class' : 'is-shown-at-desktop'
})
img_list
```

```python
# img_list에서 style 속성의 값들을 모두 출력
img = list(map(
    lambda x : x.find('div', attrs={
        'class' : 'inner'
        })['style'],
    img_list
))
img
```

```python
# title과 img 데이터를 데이터 프레임으로 변경 excel로 저장
img2 = list(map(
    lambda x : x.replace("background-image:url(", "").replace(");", ""),
    img
))
```

```python
# img2에 중복된 데이터들이 3개씩 존재한다. 중복된 데이터를 모두 삭제하고 고유한 값들만이 존재하는 리스트로 생성
img3 = []

for i in img2 :
    if i not in img3:
        img3.append(i)
img3
```

```python
# 중복된 데이터 삭제 : 집합형 데이터로 변경 -> 배열 데이터 변경
list(set(img2))
```

```python
df = pd.DataFrame({
    'title' : title,
    'img' : img3
})
df
```

```python
df.to_excel('치앙마이.xlsx', index=False)
```
